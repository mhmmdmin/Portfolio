#Metadata

1. user_id: customer account number.
2. attrition_flag: customer status (Existing and Attrited).
3. customer_age: age of the customer.
4. gender: gender of customer (M for male and F for female).
5. dependent_count: number of dependents of customers.
6. education_level: customer education level (Uneducated, High School, Graduate, College, Post-Graduate, Doctorate, and Unknown).
7. marital_status: customer's marital status (Single, Married, Divorced, and Unknown).
8. income_category: customer income interval category (Less than `$40K`, `$40K`-`$60k`, `$60K`-`$80K`, `$80K`-`$120K`, `$120K` +, and Unknown).
9. card_category: type of card used (Blue, Silver, Gold, and Platinum).
10. months_on_book: period of being a customer (in months).
11. total_relationship_count: the number of products used by customers in the bank.
12. months_inactive_12_mon: period of inactivity for the last 12 months.
13. contacts_count_12_mon: the number of interactions between the bank and the customer in the last 12 months.
14. credit_limit: credit card transaction nominal limit in one period.
15. total_revolving_bal: total funds used in one period.
16. avg_open_to_buy: the difference between the credit limit set for the cardholder's account and the current balance.
17. total_amt_chng_q4_q1: increase in customer transaction nominal between quarter 4 and quarter 1.
18. total_trans_amt: total nominal transaction in the last 12 months.
19. total_trans_ct: the number of transactions in the last 12 months.
20. total_ct_chng_q4_q1: the number of customer transactions increased between quarter 4 and quarter 1.
21. avg_utilization_ratio: percentage of credit card usage.

```python
def check_missing(df):
  missing = df.isnull().sum()
  per_mis = (missing/len(df))*100
  data_type = df.dtypes
  n_unique = df.nunique()
  return pd.DataFrame({"Missing":missing,
                       "Percent_missing_%": per_mis,
                       "Type":data_type,
                       "Unique Value": n_unique}).sort_values('Percent_missing_%', ascending=False)
```

# EDA

```python
#define function for proportion visualization
def prop(df,x):
  cross_tab_prop = pd.crosstab(index=df[x],
                             columns=df["attrition_flag"],
                             normalize="index")
  cross_tab = pd.crosstab(index=df[x],
                        columns=df["attrition_flag"])
  cross_tab_prop.plot(kind='bar', 
                        stacked=True, 
                        figsize=(10, 6))
  plt.title(f"Proportion of {x}")
  plt.legend(loc="lower left", ncol=2)
  plt.xlabel(f"{x}")
  plt.xticks(rotation = 0)
  plt.ylabel("Proportion")
  for n, x in enumerate([*cross_tab.index.values]):
    for (proportion, count, y_loc) in zip(cross_tab_prop.loc[x],
                                          cross_tab.loc[x],
                                          cross_tab_prop.loc[x].cumsum()):
      plt.text(x=n - 0.17,
               y=(y_loc - proportion) + (proportion / 2),
               s=f'{count}\n({np.round(proportion * 100, 1)}%)', 
               color="black",
               fontsize=12,
               fontweight="bold")
  plt.show()

#define function of boxplot to check column distribution
def check_distribution(df,x,y):
  sns.catplot(x = x, 
              y = y, 
              data = X_train, 
              kind='box', 
              palette='pastel')
  plt.title(f"Boxplot of {x}")
```

```python
###define function of barplot to check the relation between features###
###avg. credit limit, avg. balance used, and credit card usage###
def checkingplot(df,x):
  plt.figure(figsize=(14,6))
  #first subplot
  plt.subplot(1,3,1)
  p1 = df.groupby(x)['credit_limit'].median().plot(kind = 'bar',
                                                   ylabel='Average Credit Limit',
                                                   title=f'Average Credit Limit of {x}')
  p1.set_xticklabels(p1.get_xticklabels(), rotation=0)
  for p in p1.patches:
    p1.annotate(f'{p.get_height():0.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center',
                va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
  plt.xticks(rotation=30)

  #second subplot  
  plt.subplot(1,3,2)
  p2 = df.groupby(x)['total_revolving_bal'].median().plot(kind = 'bar',
                                                            ylabel='Average Balance Used',
                                                            title=f'Average Balance Used of {x}')
  p2.set_xticklabels(p2.get_xticklabels(), rotation=0)
  for p in p2.patches:
    p2.annotate(f'{p.get_height():0.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center',
                va = 'center',
                xytext = (0, 10),
                textcoords = 'offset points')
  plt.xticks(rotation=30)

  #third subplot    
  plt.subplot(1,3,3)
  p3 = (df.groupby(x)['avg_utilization_ratio'].mean()*100).plot(kind = 'bar',
                                                                ylabel='Credit Card Usage',
                                                                title=f'Credit Card Usage of {x}')
  p3.set_xticklabels(p3.get_xticklabels(), rotation=0)
  for p in p3.patches:
      p3.annotate(f'{p.get_height():0.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center',
                va = 'center',
                xytext = (0, 10),
                textcoords = 'offset points')
  plt.xticks(rotation=30)
  
  plt.tight_layout()
  plt.show()

#define function of pivot visualization with attrition_flag as value
def pivot(df, x, y):
  pd.pivot_table(df, index=x, columns=y, values='attrition_flag', aggfunc='mean').fillna(0).plot(kind='bar',
                                                                                                 colormap='Set3',
                                                                                                 figsize=(12,8))
  plt.xticks(rotation=0)
  plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0);
```
